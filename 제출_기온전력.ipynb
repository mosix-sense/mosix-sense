```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('~/aiffel/datathon/train.csv', encoding='cp949')

# 칼럼정보 :'num', 'date_time', '전력사용량(kWh)', '기온(°C)', '풍속(m/s)', '습도(%)','강수량(mm)', '일조(hr)', '비전기냉방설비운영', '태양광보유'

# datetime 타입으로 변환
df['date_time'] = pd.to_datetime(df['date_time'])

# 6월 1일부터 6월 15일까지의 데이터만 추출
start_date = pd.to_datetime('2020-06-01')
end_date = pd.to_datetime('2020-06-15')
df_june = df[(df['date_time'] >= start_date) & (df['date_time'] < end_date)]

# 기온과 전력 사용량 그래프 그리기
sns.set_style("white")
fig, ax2 = plt.subplots(figsize=(10, 5))

sns.lineplot(x='date_time', y='전력사용량(kWh)', data=df_june, ax=ax2, color='red', ci=0)
ax2.set_ylabel('kWh', fontsize=14, color='red')
ax2.tick_params(axis='y', colors='red')

ax1 = ax2.twinx()
sns.lineplot(x='date_time', y='기온(°C)', data=df_june, ax=ax1, color='blue', ci=0)
ax1.set_ylabel('temperature(°C)', fontsize=14, color='blue')
ax1.tick_params(axis='y', colors='blue')

plt.title('06.01 ~ 06.15', fontsize=16)
ax1.set_xlabel('date', fontsize=14)
plt.xticks(rotation=45)

plt.show()
```
