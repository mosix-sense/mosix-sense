```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
df = pd.read_csv('~/aiffel/datathon/train.csv', encoding='cp949')

#  :'num', 'date_time', '(kWh)', '(C)', '(m/s)', '(%)','(mm)', '(hr)', '', ''

# datetime  
df['date_time'] = pd.to_datetime(df['date_time'])

# 6 1 6 15  
start_date = pd.to_datetime('2020-06-01')
end_date = pd.to_datetime('2020-06-15')
df_june = df[(df['date_time'] >= start_date) & (df['date_time'] < end_date)]

#     
sns.set_style("white")
fig, ax2 = plt.subplots(figsize=(10, 5))

sns.lineplot(x='date_time', y='(kWh)', data=df_june, ax=ax2, color='red', ci=0)
ax2.set_ylabel('kWh', fontsize=14, color='red')
ax2.tick_params(axis='y', colors='red')

ax1 = ax2.twinx()
sns.lineplot(x='date_time', y='(%)', data=df_june, ax=ax1, color='green', ci=0)
ax1.set_ylabel('humidity(%)', fontsize=14, color='green')
ax1.tick_params(axis='y', colors='green')

plt.title('06.01 ~ 06.15', fontsize=16)
ax1.set_xlabel('date', fontsize=14)
plt.xticks(rotation=45)

plt.show()
```
